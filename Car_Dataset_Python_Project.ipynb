```python
data=pd.read_csv("/content/drive/MyDrive/data.csv")
```

```python
df=pd.DataFrame(data)
df
```

```python
#Engine Fuel Type column filling null values with the most frequent value
df["Engine Fuel Type"].fillna(df["Engine Fuel Type"].mode()[0],inplace=True)
df["Engine Fuel Type"].isna().sum()

```

```python
#Engine HP column filling null values with mean
df["Engine HP"].fillna(df["Engine HP"].mean(),inplace=True)
df["Engine HP"].isna().sum()
```

```python
#Engine Cylinders column filling null values with mean
df["Engine Cylinders"].fillna(df["Engine Cylinders"].mean(),inplace=True)
df["Engine Cylinders"].isna().sum()
```

```python
#Number of Doors column filling null values with mean
df["Number of Doors"].fillna(df["Number of Doors"].mean(),inplace=True)
df["Number of Doors"].isna().sum()
```

```python
#Market Category column filling null values with mode
df["Market Category"].fillna(df["Market Category"].mode()[0],inplace=True)
df["Market Category"].isna().sum()
```

```python
#Filter the dataset to only include cars from the year 1995 and later
df=df[df["Year"]>=1995]
```

```python
#Standardize text entries by converting Vehicle Style column entries to lowercase
df["Vehicle Style"]=df["Vehicle Style"].str.strip().str.lower()
```

```python
#Standardize text entries by converting Market Category column entries to lowercase
df["Market Category"]=df["Market Category"].str.strip().str.lower()
```

```python
#Create a column called Total MPG that is the average of city mpg and highway MPG.
df["Total MPG"]=(df["city mpg"]+df["highway MPG"])/2
```

```python
#Create a column called Price per HP calculated as:
df["Price per HP"]=df["MSRP"] / df["Engine HP"]
```

```python
#Calculate summary statistics (mean, median, standard deviation) for the following columns:
#Engine HP, MSRP, Popularity, highway MPG, and city mpg
df[["Engine HP","MSRP","Popularity","highway MPG","city mpg"]].describe()
```

```python
#Group the data by the following columns and calculate the average MSRP and Popularity for each group:
#Driven_Wheels
#Vehicle Size
#Engine Cylinders
grouped_df=df.groupby(["Driven_Wheels","Vehicle Size","Engine Cylinders"]).agg({"MSRP":"mean","Popularity":"mean"})
grouped_df
```

```python
#A histogram that shows a distribution for the city mpg column.
plt.figure(figsize=(10, 6))
plt.hist(df["city mpg"],bins=12,edgecolor="black")
plt.xlabel("city mpg")
plt.ylabel("Frequency")
plt.title("Distribution of city mpg")
plt.show()
```

```python
#A bar chart showing the average MSRP for each category in Vehicle Size.
plt.figure(figsize=(10, 6))
bar_chart=df.groupby("Vehicle Size")["MSRP"].mean().plot(kind="bar",color="skyblue")
plt.xlabel("Vehicle Size")
plt.ylabel("Average MSRP")
plt.title("Average MSRP by Vehicle Size")
plt.xticks(rotation=0)
plt.show()
```

```python
#A scatter plot showing the relationship between Engine HP and MSRP.
plt.figure(figsize=(10, 6))
scatter_plot=plt.scatter(data=df,x="Engine HP",y="MSRP",color="green",alpha=0.7)
plt.xlabel("Engine HP")
plt.ylabel("MSRP")
plt.title("Relationship between Engine HP and MSRP")
plt.show()
```

```python
#A boxplot showing the distribution of MSRP for each category in Driven_Wheels.
plt.figure(figsize=(10, 6))
boxplot=sns.boxplot(data=df,x="Driven_Wheels",y="MSRP",hue="Driven_Wheels",palette="Set2")
plt.xlabel("Driven Wheels")
plt.ylabel("MSRP")
plt.title("Distribution of MSRP by Driven_Wheels")
plt.xticks(rotation=45)
plt.show()
```

```python
#A line plot showing trends in average city mpg and highway MPG for different Transmission Type
plt.figure(figsize=(10, 6))
line_plot=df.groupby("Transmission Type")[["city mpg","highway MPG"]].mean()
plt.plot(line_plot,marker="o",linewidth=2)
plt.xlabel("Transmission Type")
plt.ylabel("Average MPG")
plt.title("Trends in Average MPG by Transmission Type")
plt.legend(["city mpg","highway MPG"])
plt.show()
```

```python
#Investigate the correlation between the following variables:
#Engine HP, MSRP, Popularity, city mpg, and highway MPG
correlation_matrix=df[["Engine HP","MSRP","Popularity","city mpg","highway MPG"]].corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix,annot=True,cmap="coolwarm",fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
